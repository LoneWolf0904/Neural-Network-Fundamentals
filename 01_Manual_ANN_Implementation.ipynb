```python
import numpy as np

# Initialising neural network as per the given parameters

ip_neurons = 2
hidden_neurons = 2
op_neurons = 1

alpha = 0.1
Epochs = 10000

# Defining Sigmoid Function and derivative

def sigmoid(x):
  return 1/(1+np.exp(-x))

def sigmoid_der(x):
  return x*(1-x)

# NPC training data

X = np.array([[1,1], [0,1], [1,0], [0,0]])

# NPC Output data

Y = np.array([[0], [1], [1], [0]])

# Randomizing initial weights and biases

w_input_hidden = np.random.uniform(size=(ip_neurons, hidden_neurons))
bias_hidden = np.zeros((1, hidden_neurons))

w_hidden_output = np.random.uniform(size=(hidden_neurons, op_neurons))
bias_output = np.zeros((1, op_neurons))

# ANN Training

for Epoch in range(Epochs):

  # Feed Forward

  hidden_input = np.dot(X, w_input_hidden) + bias_hidden
  hidden_output = sigmoid(hidden_input)

  output_layer_input = np.dot(hidden_output, w_hidden_output) + bias_output
  Ycurrent = sigmoid(output_layer_input)

  # Calculating Error

  e = Y - Ycurrent

  # Backpropagation

  op_error = e * sigmoid_der(Ycurrent)
  hidden_error = op_error.dot(w_hidden_output.T) * sigmoid_der(hidden_output)

  # Updating weights and biases

  w_hidden_output += hidden_output.T.dot(op_error) * alpha
  bias_output += np.sum(op_error, axis = 0, keepdims = True) * alpha

  w_input_hidden += X.T.dot(hidden_error) * alpha
  bias_hidden += np.sum(hidden_error, axis = 0, keepdims = True) * alpha

# Testing the Neural Network

test_data = np.array([[1,1], [0,1], [1,0], [0,0]])

hidden_input = np.dot(test_data, w_input_hidden) + bias_hidden
hidden_output = sigmoid(hidden_input)

output_layer_input = np.dot(hidden_output, w_hidden_output) + bias_output
Ycurrent = sigmoid(output_layer_input)

# Displaying the ANN predictions

for i in range(len(test_data)):
  dec = 1 if Ycurrent[i] > 0.5 else 0

  print(f"NPC's power: {test_data[i][0]}, Enemy's power: {test_data[i][1]}, Decision: {'Attack' if dec == 0 else 'Flee'}")
```
